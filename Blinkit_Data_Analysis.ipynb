```python
total_sales=df['Sales'].sum()
print(f"Total Sales:${total_sales:,.0f}")
```

```python
avg_sales = df['Sales'].mean()
print(f"Average Sales: ${avg_sales}")
```

```python
no_of_items_sold = df['Sales'].count()
print(f'Number of items sold: {no_of_items_sold}')
```

```python
avg_rating = df['Rating'].mean()
print(f"Average Rating: {avg_rating:,.0f}")
```

```python
sales_by_fat=df.groupby(['Item Fat Content'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
ax = sns.barplot(x='Item Fat Content',y='Sales',data=sales_by_fat)
plt.title('Sales by Item Fat Content')

for container in ax.containers:
  ax.bar_label(container)
```

```python
sales_by_fat=df.groupby(['Item Fat Content'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
plt.pie(sales_by_fat['Sales'], labels=sales_by_fat['Item Fat Content'], autopct='%.2f%%', startangle=90)
plt.title('Sales by Item Fat Content')
plt.show()
```

```python
sales_by_Item_Fat_Content=df.groupby(['Item Fat Content'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)

bars = sns.barplot(x='Item Fat Content',y='Sales',data=sales_by_Item_Fat_Content)
plt.title('Sales by Item Fat Content')
sns.set(rc={'figure.figsize':(5,3)})

for container in bars.containers:
  bars.bar_label(container, fmt='%.0f')



plt.xticks(rotation=90)
plt.xlabel('Item Fat Content')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
sales_by_Item_Fat_Content=df.groupby(['Item Fat Content'],as_index=False)['Sales'].sum().sort_index()
plt.figure(figsize=(5,5))
plt.plot(sales_by_Item_Fat_Content['Item Fat Content'], sales_by_Item_Fat_Content['Sales'], marker='o', linestyle='-')
plt.title('Sales by Item Fat Contentr')

plt.xticks(rotation=90)
plt.xlabel('Item Fat Content')
plt.ylabel('Sales')

for x,y in zip(sales_by_Item_Fat_Content['Item Fat Content'], sales_by_Item_Fat_Content['Sales']):
  plt.text(x,y,f'${y:,.0f}', ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
sales_by_type=df.groupby(['Item Type'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)

bars = sns.barplot(x='Item Type',y='Sales',data=sales_by_type)
plt.title('Sales by Item Type')
sns.set(rc={'figure.figsize':(10,8)})

for container in bars.containers:
  bars.bar_label(container, fmt='%.0f')



plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```

```python
sales_by_outlet=df.groupby(['Outlet Location Type', 'Item Fat Content'],as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x='Outlet Location Type',y='Sales',data=sales_by_outlet, hue='Item Fat Content')
plt.title('Sales by Outlet Location Type and Item Fat Content')
```

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

ax= grouped.plot(kind='bar', figsize=(8,5))
plt.xlabel('Outlet Location Type')
plt.ylabel('Sales')
plt.title('Sales by Outlet Location Type and Item Fat Content')

ax=plt.gca()
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```
